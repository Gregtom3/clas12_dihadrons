```python
import ROOT
import copy
```

```python
def get_pi0_pi0_dist(rootname,title,cut):
    tfile = ROOT.TFile(f"/work/clas12/users/gmat/clas12/clas12_dihadrons/projects/ana_v1/volatile/data/pi0_pi0/{rootname}")
    tree = tfile.Get("dihadron")
    #print("\nTotal pi0pi0 dihadrons =",tree.GetEntries())
    #print("Total true pi0pi0 dihadrons =",tree.GetEntries("trueparentpid_1==111&&trueparentpid_2==111"))
    # Create the TH2F histogram
    xmin = ymin = 0.005
    xmax = ymax = 0.4
    nbins = 100
    histogram = ROOT.TH2F("histogram", title+";M_{1} (GeV);M_{2} (GeV)", nbins, xmin, xmax, nbins, ymin, ymax)
    histogram.GetXaxis().SetNdivisions(5)
    histogram.GetYaxis().SetNdivisions(5)

    # Set the log scale on the z-axis
    ROOT.gStyle.SetOptStat(0)
    ROOT.gStyle.SetPalette(ROOT.kRainBow)
    ROOT.gStyle.SetPaintTextFormat("4.1f")
    ROOT.gStyle.SetNumberContours(999)
    ROOT.gStyle.SetPadRightMargin(0.16)

    tree.Draw("M2:M1>>histogram",cut,"")

    # Create and draw the canvas
    canvas = ROOT.TCanvas("canvas", "canvas", 800, 600)
    histogram.GetZaxis().SetTitleOffset(1.2)
    histogram.GetZaxis().SetRangeUser(1, histogram.GetMaximum())
    histogram.GetZaxis().SetTitle("Counts")
    histogram.Draw("colz")
    canvas.SetLogz()

    # Show the plot
    return copy.deepcopy(canvas)
```

```python
canvas=get_pi0_pi0_dist("MC_RGA_inbending_merged.root","RGA Monte Carlo #pi^{0}#pi^{0}","")
canvas.Draw()
```
