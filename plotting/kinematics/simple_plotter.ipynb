`plot_graph(**args)`
- `project_dir`: The path to the project directory.
- `channel`: The channel being analyzed (e.g. piplus_piminus, piminus_piminus, pi0_pi0).
- `version`: The version of the data being analyzed (e.g. MC_RGA_inbending, Data_RGC).
- `ttree`: The name of the ROOT TTree containing the data to be plotted.
- `x_col`: The name of the column to be plotted on the x-axis.
- `x_min`: The minimum value for the x-axis range.
- `x_max`: The maximum value for the x-axis range.
- `x_bins`: The number of bins to use for the x-axis histogram. If not specified, defaults to 50.
- `y_col`: The name of the column to be plotted on the y-axis. If not specified, a 1-dimensional histogram is created.
- `y_min`: The minimum value for the y-axis range.
- `y_max`: The maximum value for the y-axis range.
- `y_bins`: The number of bins to use for the y-axis histogram. If not specified, defaults to 50.
- `set_log_y`: Set to `True` to use a logarithmic scale for the y-axis.
- `z_min`: The minimum value for the z-axis range (for 2D histograms).
- `z_max`: The maximum value for the z-axis range (for 2D histograms).
- `z_col`: The name of the column to be used for the z scaling
- `set_log_z`: Set to `True` to use a logarithmic scale for the z-axis (for 2D histograms).
- `cut`: The selection criteria to apply to the data.
- `nbins`: The default number of bins to use for histograms.
- `latex_xmax`: The x-coordinate for the top-right corner of the title text.
- `latex_ymax`: The y-coordinate for the top-right corner of the title text.
- `alias_x`: An alias for the x-axis column name.
- `alias_y`: An alias for the y-axis column name.
- `alias_z`: An alias for the z-axis column name.
- `suffix`: A suffix to add to the filename for saving the plot. If not specified, the plot is not saved to a file.
