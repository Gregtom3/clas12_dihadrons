```python
import ROOT
```

```python

f1=ROOT.TFile("old_5032.root")
t1=f1.Get("dihadron")

f2=ROOT.TFile("new_5032.root")
t2=f2.Get("dihadron")

print("Old nSidis 5032 PiPlusPi0\t nEntries=",t1.GetEntries())
print("New nSidis 5032 PiPlusPi0\t nEntries=",t2.GetEntries())
```

```python
# Create a canvas and set its title
c = ROOT.TCanvas("c", "Histograms")

# Draw the first histogram and set its line color to red
t1.Draw("phi_h>>h1(50,-3.14,3.14)")
h1 = ROOT.gDirectory.Get("h1")
h1.SetLineColor(ROOT.kRed)

# Draw the second histogram on the same canvas and set its line color to black
t2.Draw("phi_h>>h2(50,-3.14,3.14)", "", "same")
h2 = ROOT.gDirectory.Get("h2")
h2.SetLineColor(ROOT.kBlack)

# Set the titles and labels of the axes and the legend
h1.GetXaxis().SetTitle("#phi_{h}")
h1.GetYaxis().SetTitle("Counts")
h1.SetTitle("Histograms of #phi_{h}")
legend = ROOT.TLegend(0.6, 0.7, 0.9, 0.9)
legend.AddEntry(h1, "t1", "l")
legend.AddEntry(h2, "t2", "l")

# Draw the legend and the canvas
legend.Draw()
c.Draw()
```
